```python
import pandas as pd
df = pd.read_csv('https://query.data.world/s/qfftbmw36gqeexuztx5gla6t3urpo3')
df.head()
```

```python
#dropping fields not needed for analyses to improve efficiency
del df['uid']
```

```python
#sorting date column to find earliest listed date here
df = df.sort_values(by="date")
df.head()
```

```python
#using describe function to see if there are any anomalies in the data
df['fips_code'].describe()
```

```python
#The fips code 99999 does not exist. Sorting to see what this is.
df = df.sort_values(by="fips_code", ascending=False)
df.head()
```

```python
max_fips = 78030
covid_data = df[(df["fips_code"] <= max_fips)]
covid_data = covid_data.sort_values(by="fips_code", ascending=False)
covid_data.head()
```

```python
#other dataset does not have the fips code so we will delete it here
del covid_data['fips_code']
```

```python
#date format is different in pollutant csv so changing format here to match
covid_data['date'] = pd.to_datetime(covid_data['date'])
covid_data.head()
```

```python
#using describe function to see if there are any anomalies in the data
covid_data['total_population'].describe()
```

```python
#using describe function to see if there are any anomalies in the data
covid_data['cumulative_cases_per_100_000'].describe()
```

```python
#converting the df to a csv if we want to merge and create the database in postgres
covid_data.to_csv('./covid_part1.csv')
```
