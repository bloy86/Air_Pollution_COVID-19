```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
cv19 = pd.read_csv("covid_part1.csv") 
pol = pd.read_csv("pollutant_data.csv")
```

```python
c_cv19= cv19[['state', 'location_name','date','total_population', 'cumulative_cases','cumulative_deaths','new_cases',
'new_deaths']]
```

```python
c_cv19
```

```python
c_pol = pol.drop('Unnamed: 0',1)
```

```python
merged_df = c_pol.merge(c_cv19, how='inner', left_on=["State","County","Date"], right_on=["state", "location_name","date"])
```

```python
cvpo = merged_df.drop(['state','location_name','date'], axis = 1)
```
